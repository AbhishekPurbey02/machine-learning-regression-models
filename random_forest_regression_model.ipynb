```python
# Importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
# Importing the datasets
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
# Training the Random forest regression model
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(x,y)
```

```python
regressor.predict([[6.5]])
```

```python
# Visualising the result of Random forest regressor
plt.scatter(x,y, color = 'red')
plt.plot(x,regressor.predict(x), color = 'blue')
plt.title('Truth or Bluff(Random Forest Regression )')
plt.xlabel('Position Label')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Random Forest Regression with higher resolution results 
x_gird = np.arange(min(x), max(x), 0.01)
x_gird = x_gird.reshape((len(x_gird), 1))
plt.scatter(x,y, color = 'red')
plt.plot(x_gird, regressor.predict(x_gird), color = 'blue')
plt.title('Truth or Bluff(Random forest regressor)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
